```python
import osmnx as ox
import pandas as pd
import cartopy.crs as ccrs
import cartopy.io.shapereader as shpreader
import matplotlib.pyplot as plt
import cartopy.feature as cf
```

```python
cr = ox.graph_from_place("Costa Rica", network_type='all_private')
```

```python
ox.save_graph_geopackage(cr, filepath="cr_allPrivate.gpkg")
```

```python
fig, ax = ox.plot_graph(cr)
```

```python
nodes, streets = ox.graph_to_gdfs(cr)
```

```python
nodes.to_file("nodesCR")
```

```python
nodes.head()
```

```python
land = cf.NaturalEarthFeature(
    category='physical',
    name='land',
    scale='10m',
    facecolor=cf.COLORS['land'],
    alpha=0.5)
```

```python
figure = plt.figure(figsize=(20,20))
ax = plt.axes(projection = ccrs.LambertConformal())


ax.set_extent([-84.313711, -84.232153, 10.412284,10.492346 ])

land = cf.NaturalEarthFeature(
    category='physical',
    name='land',
    scale='10m',
    facecolor=cf.COLORS['land'],
    edgecolor='black',
    alpha=0.5)

ax.add_geometries(streets.geometry, crs=ccrs.PlateCarree(), zorder=5, edgecolor='blue')

ax.add_feature(land)

ax.gridlines()
```
